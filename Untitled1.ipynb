```python
import pandas as pd
import matplotlib.pyplot as plt

## PART 1

url = r"C:\Users\Marie-Laure\Desktop\amz_uk_price_prediction_dataset.csv"

df = pd.read_csv(url)

freq_category = df["category"].value_counts()
freq_category = freq_category.reset_index()
freq_category = freq_category.sort_values(by='count', ascending=False)
freq_category = freq_category[[idx < 5 for idx, y in freq_category.iterrows()]]
freq_category

plt.figure(figsize=(10,6))
plt.bar(freq_category['category'], freq_category['count'], color='red')
plt.xticks(rotation=45)

for idx, row in freq_category.iterrows():
    plt.text(
        x = idx,
        y = row['count'] / 2,
        s = row['count'],
        ha = 'center')
plt.show()
```
